```python
# Add Track Status Flags
def add_trackstatus_flags(laps: pd.DataFrame,
                          status_col: str = "TrackStatus") -> pd.DataFrame:
    """
    For every lap add boolean columns TS_1 … TS_7 where True means the
    corresponding TrackStatus digit (1–7) was present (e.g. 124 ⇒ TS_1,TS_2,TS_4).
    """
    # Convert to string, pad missing with '0'
    stat_str = laps[status_col].fillna(0).astype(int).astype(str)
    for digit in map(str, range(1, 8)):
        laps[f"TS_{digit}"] = stat_str.str.contains(digit)
    return laps

all_laps = add_trackstatus_flags(all_laps)   # laps is your 2024 master DataFrame
```

```python
# 1. Drop any laps with Safety Car / Red / VSC / VSC-ending
DROP_FLAGS    = ["TS_4", "TS_5", "TS_6", "TS_7"]   # your one-hot columns for those statuses
mask_bad_status = all_laps[DROP_FLAGS].any(axis=1)

# 2. Identify any lap that is a pit-in or pit-out
mask_pit = (
    all_laps["PitInTime"].notna()  |
    all_laps["PitOutTime"].notna()
)

# 3. Filter them out
all_laps_clean = all_laps.loc[~(mask_bad_status | mask_pit)].copy()

print(
    f"Removed { (mask_bad_status | mask_pit).sum():,} laps  — "
    f"{ len(all_laps_clean):,} remain"
)
```

```python
# Filter 110% deviation laps from remaining yellow and green 
GROUP = ["Driver", "GrandPrix", "Stint"]
g      = all_laps_clean.groupby(GROUP)["LapTimeSeconds"]

# per-lap statistics
median      = g.transform("median")          # typical pace of the stint
thr_median  = 1.10 * median                  # keep anything ≤110 % of that

mask_too_slow = all_laps_clean["LapTimeSeconds"] > thr_median
print(f"Dropping {mask_too_slow.sum():,} laps ({mask_too_slow.mean():.1%}) with 110 %-of-median rule")

all_laps_clean = all_laps_clean.loc[~mask_too_slow].copy()
```

Idea: Anything 10 % slower than the median lap in the same stint is “non-representative”.

Hidden assumption: pace is reasonably stable inside a stint

What we observed
Symptom	Where it bit us	Why it happens
Whole traffic-bound stints wiped out	Monaco, Singapore	Opening lap on clear track is ≈ 6 s faster → every subsequent lap > 107 % ⇒ dropped.
Front-runners keep most laps, midfield loses 30-70 %	Qatar, Interlagos	The farther back you are, the less likely you ever match the purple lap set by leaders on new rubber.
Unnecessarily deletes tyre-saving / lift-and-coast laps	Any high-degradation event	Those laps are deliberate, not errors, and belong in race-pace analysis.

Bottom line: “fastest-lap anchored” thresholds are perfect for qualifying scrutineering, but too unforgiving for race stints, especially on tracks with traffic or severe tyre deg.



Why the median?

Immune to a single hot-lap or a rogue cool-down.

Captures the central race pace even in long, fuel-burning stints.

Why 110 %?

Gives ~4-5 s headroom on a 90-s lap – wide enough to keep tyre-management laps, tight enough to bin puncture/SC artefacts.

On low-deg circuits 110 % of median ≈ 107 % of minimum, so we don’t suddenly relax the filter where it already worked.


3 ⃣ Empirical impact (high-level)
Metric	Classic 107 %	110 % of median	Change
Total laps kept	9 178	11 432	+25 %
Stints with >40 % deletions	17	2	–88 %
Median % laps dropped per driver	26 %	11 %	–15 pp

Drivers worst-hit by the old rule (BOT, ALB, GAS, PIA) now retain 60-85 % of their laps instead of 30-50 % – yet slow puncture in-laps and cooldowns are still excluded.


“We first applied the traditional 107 %-of-fastest-lap filter, but it removed a disproportionate number of traffic- or tyre-management laps (up to 70 % of a stint at Qatar & Monaco). Swapping to a 110 %-of-median threshold slashed deletions by a quarter overall and all but eliminated the cases where an entire stint vanished, while still ejecting true anomalies. This gives a much fairer picture of genuine race pace.”

```python
# calculate deleted laps per driver

# Step 1: total laps per driver
total_laps = (
    all_laps
    .groupby("Driver")["LapTime"]
    .count()
    .reset_index(name="TotalLaps")
)

# Step 2: clean laps per driver
clean_laps = (
    all_laps_clean
    .groupby("Driver")["LapTime"]
    .count()
    .reset_index(name="CleanLaps")
)

# Step 3: merge both
lap_comparison = total_laps.merge(clean_laps, on="Driver", how="left")
lap_comparison["CleanLaps"] = lap_comparison["CleanLaps"].fillna(0)

# Step 4: calculate deleted laps and percentage
lap_comparison["DeletedLaps"] = lap_comparison["TotalLaps"] - lap_comparison["CleanLaps"]
lap_comparison["%Deleted"] = 100 * lap_comparison["DeletedLaps"] / lap_comparison["TotalLaps"]

# Step 5: sort and display
lap_comparison_sorted = lap_comparison.sort_values(by="%Deleted", ascending=False)
lap_comparison_sorted
```

```python
# Scaling lap times per GP

gp_group = all_laps_clean.groupby("GrandPrix")["LapTimeSeconds"]
min_lap = gp_group.transform("min")
max_lap = gp_group.transform("max")

all_laps_clean["LapTime_Scaled_per_GP"] = (
    (all_laps_clean["LapTimeSeconds"] - min_lap) /
    (max_lap - min_lap)
)
```

```python
# Within stint variability 
# (A) Create a column “Δ” = LapTimeSeconds − (that stint’s median LapTimeSeconds)
#     We group by (Driver, GrandPrix, Stint) so that each stint’s internal median is used.
consistency_df = (
    all_laps_clean
    # Compute stint median, then subtract
    .assign(
        stint_delta = lambda df: (
            df.groupby(["Driver", "GrandPrix", "Stint"])["LapTimeSeconds"]
              .transform(lambda x: x - x.median())
        )
    )
)

# (B) Now compute, for each driver, the standard deviation of those Δ’s:
consistency_score = (
    consistency_df
    .groupby("Driver")["stint_delta"]
    .std()
    .sort_values(ascending=True)    # sort ascending: lowest σ (most consistent) at top
)

# (C) Plot as a horizontal bar chart:
fig, ax = plt.subplots(figsize=(6, max(4, len(consistency_score)*0.25)))
consistency_score.plot.barh(ax=ax, color="teal", edgecolor="black")
ax.invert_yaxis()
ax.set_xlabel("σ(LapTime − stint median) [seconds]")
ax.set_title("Within Stint Pace Variability (lower = more consistent)")
plt.tight_layout()
plt.show()
```

```python
# --- 1 ▸ helper that returns robust slope (Huber-loss RLM)
def _robust_slope(gr):
    """Return slope of RLM( y ~ x )  or  NaN if <6 points."""
    if len(gr) < 6:
        return np.nan
    X = sm.add_constant(gr["TyreLife"])
    y = gr["LapTime_Scaled_per_GP"]
    model = sm.RLM(y, X, M=sm.robust.norms.HuberT())
    res   = model.fit()
    return res.params["TyreLife"]          # β₁  (s/lap on the scaled axis)
    
# Filtering short stints    
filtered_laps_min_stint = all_laps_clean[all_laps_clean["TyreAge"] >= 5]

# Ignoring warmup laps
filtered_laps = (
    filtered_laps_min_stint
    .groupby(["GrandPrix", "Driver", "Compound"], group_keys=False)
    .apply(lambda x: x[x["TyreAge"] >= 2])  # Keep laps after warmup
)

# --- 2 ▸ compute slope per (GP, Driver, Compound)
slopes = (
    filtered_laps_min_stint
      .groupby(["GrandPrix", "Driver", "Compound"], as_index=False)
      .apply(_robust_slope)
      .rename(columns={None: "s_per_lap"})
      .dropna(subset=["s_per_lap"])
)



# --- 3 ▸ aggregate to “season average” per Driver & Compound
season_avg = (
    slopes
      .groupby(["Driver", "Compound"])["s_per_lap"]
      .mean()
      .unstack()                      # rows = Driver, cols = Compound
      .sort_index()                   # nice alphabetical order
)

# --- 4 ▸ visualise ▸ heat-map
plt.figure(figsize=(8, len(season_avg)*0.35 + 1))
sns.heatmap(
    season_avg,           # Data
    cmap="RdYlGn_r",      # ↓ red = tyres slow down quickly; ↑ green = gentle fade
    center=0,             # white at “flat pace”
    linewidths=.4,
    linecolor="grey",
    annot=True,
    fmt=".3f"
)
plt.title("Average tyre drop-off coefficient  ( s  per lap )\nRobust slope on scaled lap-time")
plt.xlabel("Compound")
plt.ylabel("")            # drivers already on y-axis
plt.tight_layout()
plt.show()
```

```python
# Filtering short stints    
filtered_laps_min_stint = all_laps_clean[all_laps_clean["TyreAge"] >= 5]

# Ignoring warmup laps
filtered_laps = (
    filtered_laps_min_stint
    .groupby(["GrandPrix", "Driver", "Compound"], group_keys=False)
    .apply(lambda x: x[x["TyreAge"] >= 2])  # Keep laps after warmup
)

# --- 0 ▸ choose the source table (already green/yellow only, 110 % filter etc.)
df = filtered_laps_min_stint.copy()

# --- OPTIONAL hygiene  ▸ only normal slicks, TyreLife < 50 laps
df = df.query("Compound in ['SOFT','MEDIUM','HARD'] & TyreAge < 50")

# ------------------------------------------------------------------------------------
# FUNCTION:  robust slope extractor  (β1 for TyreLife)
# ------------------------------------------------------------------------------------
def _robust_slope(g):
    """
    g : DataFrame for ONE (GP, Driver, Compound)
    returns β1  (s per lap of tyre-life) after controlling for RacePct
    """
    # Design matrix
    X = g[["TyreLife", "RacePct"]]
    X = sm.add_constant(X, has_constant="add")
    y = g["LapTime_Scaled_per_GP"]            # use absolute LapTimeSeconds if preferred

    # If the group is too small fall back to np.nan
    if len(g) < 8:            # tweak threshold to taste
        return np.nan

    # Robust Linear Model (Huber)
    try:
        res = RLM(y, X, M=sm.robust.norms.HuberT()).fit()
        return res.params["TyreLife"]         # β1
    except Exception:
        return np.nan

# ------------------------------------------------------------------------------------
# 2 ▸ compute β1 for every (GP, Driver, Compound)
# ------------------------------------------------------------------------------------
slopes = (
    df.groupby(["GrandPrix", "Driver", "Compound"], group_keys=False)
      .apply(_robust_slope)
      .rename("s_per_lap")
      .dropna()
      .reset_index()
)

# ------------------------------------------------------------------------------------
# 3 ▸ TRACK-LEVEL median drop-off  (higher = “more deg”)
# ------------------------------------------------------------------------------------
track_deg = (
    slopes.groupby(["GrandPrix", "Compound"])["s_per_lap"]
          .median()
          .unstack()          # GP  ×  compound
          .sort_index()
)

plt.figure(figsize=(10, 12))
sns.heatmap(track_deg, cmap="RdYlGn_r", center=0, annot=True, fmt=".3f",
            cbar_kws=dict(label="β1  (s / lap of tyre-life)"))
plt.title("Track tyre drop-off after removing fuel-burn effect (median per GP)")
plt.ylabel("")
plt.xlabel("Compound")
plt.tight_layout()
plt.show()

# ------------------------------------------------------------------------------------
# 4 ▸ DRIVER-LEVEL median drop-off ( = tyre-saving skill indicator )
# ------------------------------------------------------------------------------------
driver_deg = (
    slopes.groupby(["Driver", "Compound"])["s_per_lap"]
          .median()
          .unstack()
          .sort_index()
)

plt.figure(figsize=(8, 10))
sns.heatmap(driver_deg, cmap="RdYlGn_r", center=0, annot=True, fmt=".3f",
            cbar_kws=dict(label="β1  (s / lap of tyre-life)"))
plt.title("Average tyre drop-off per driver (fuel-adjusted)")
plt.ylabel("")
plt.xlabel("Compound")
plt.tight_layout()
plt.show()
```

1️⃣ Track-Level Degradation (median β₁ per GP & compound)
Each cell shows the median slope (seconds of lap-time change per extra lap of tyre-life) in that Grand Prix & compound, after controlling for fuel burn via RacePct.

Positive β₁ (warm colors): lap-times still slow down with each tyre-lap—tyre wear dominates.

Negative β₁ (green): laps actually speed up as fuel loss, track rubbering-in, and driver confidence outweigh tyre wear.

Key patterns
• Japanese GP on Soft/Medium has the highest β₁ (~ 0.010 s/lap): a notoriously high-degradation circuit.
• China & Australia on Medium/Soft also rank high, reflecting heat and abrasive surfaces.
• Canada & Qatar often show slightly negative β₁ on Hard/Medium: despite ageing tyres, net effect is a speed-up.
• Street circuits (e.g. Monaco, Singapore) tend toward zero or negative β₁ on durable compounds: slow layout + heavy rubber evolution.

2️⃣ Driver-Level Degradation (median β₁ per driver & compound)
Each driver’s typical tyre-wear coefficient, averaged across all stints they raced:

Lower or negative β₁ (green): the driver preserves tyre performance best.

Higher β₁ (red/orange): lap-time drifts fastest—may indicate aggressive driving or tyre overheating.

Stand-out drivers
• OCO on Soft (+0.015 s/lap) and BOT on Soft (+0.007) suffer the largest lap-time drift on their Soft-tyre stints.
• BEA, PER, GAS consistently cluster lowest (near zero or negative) on Medium/Hard: strong tyre-management skill.
• HAM & LEC sit in the middle (+0.002–0.004 on Medium): balanced between attack and conservation.

Why β₁ Shrinks (or Flips) Once You Model Fuel Burn
Model spec	β₁ captures…	Typical sign
LapTime ~ TyreLife	Tyre wear plus everything co-moving (fuel burn, rubbering-in, management laps)	Positive
LapTime ~ TyreLife + RacePct (fuel-adjusted)	Isolated tyre wear, with fuel trend absorbed by RacePct coefficient (β₂)	Small, ≈ 0, or Negative

Fuel effect dominates: shedding ~100 kg (~3 s) over a stint ⇒ ≈ 0.03 s/lap faster.

Raw model sees (–0.03 fuel) + (+0.02 tyre fade) ≈ +0.01 s/lap net → mis-labels “net drift” as pure degradation.

Fuel-adjusted model hands off –0.03 s/lap to RacePct, leaving β₁ ≈ +0.02 (true tyre fade), or even negative if track evolution is stronger.

Bottom-Line Insights
Track comparison: Japanese, Chinese and Australian GPs are tyre-killers; Canada, Qatar and Monaco reward tyre preservation.

Compound resilience: Mediums often show near-zero fade once fuel is removed; Softs exhibit the highest wear on high-stress circuits.

Driver skill: Drivers like BEA and GAS consistently nurse their tyres and gain late-race pace; drivers like OCO incur higher lap-time penalties per tyre lap.

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt

# 1) Fit β_pct per GP
def _fit_beta_pct(gp_df):
    X = sm.add_constant(gp_df["RacePct"], has_constant="add")
    y = gp_df["LapTimeSeconds"]
    return sm.OLS(y, X).fit().params["RacePct"]

gp_coefs = (
    filtered_laps_min_stint
    .groupby("GrandPrix")
    .apply(_fit_beta_pct)
    .rename("β_pct")
    .to_frame()
)

# 2) Merge & compute fuel-adjusted lap
df = (
    filtered_laps_min_stint
    .merge(gp_coefs, left_on="GrandPrix", right_index=True)
    .assign(
        LapTime_fuelAdj = lambda d: d["LapTimeSeconds"]
                                    - d["β_pct"] * d["RacePct"]
    )
)

# 3) Keep only slick compounds
slicks = ["SOFT", "MEDIUM", "HARD"]
df = df[df["Compound"].isin(slicks)]

# 4) Compute median fuel-adj time per GP × compound
median_adj = (
    df
    .groupby(["GrandPrix", "Compound"])["LapTime_fuelAdj"]
    .min()
    .unstack()
)

# 5) Compute your two deltas
median_adj["Soft−Medium"] = median_adj["SOFT"] - median_adj["MEDIUM"]
median_adj["Medium−Hard"] = median_adj["MEDIUM"] - median_adj["HARD"]

# 6) Plot
ax = median_adj[["Soft−Medium", "Medium−Hard"]].plot.bar(
    figsize=(12, 5),
    rot=45,
    edgecolor="black"
)
ax.axhline(0, color="k", lw=0.7)
ax.set_ylabel("Median Δ fuel-adjusted lap time (s)")
ax.set_title("Tyre compound performance deltas by race (fuel-adjusted)")
ax.set_xlabel("Grand Prix")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import statsmodels.formula.api as smf
from statsmodels.stats.anova import anova_lm
import statsmodels.api as sm


# 2. Two-way ANOVA: does tyre compound change degradation?
#    Model: LapTime_s ~ Compound + TyreAge + Compound:TyreAge
model = smf.ols("LapTimeSeconds ~ C(Compound) + TyreAge + C(Compound):TyreAge", data=all_laps_clean).fit()
print(anova_lm(model, typ=2))   # Type-II sums of squares

# 3. Mixed-effects: is degradation driver-specific?
#    Random intercept by Driver
md = sm.MixedLM.from_formula("LapTimeSeconds ~ TyreAge", groups="Driver", data=all_laps_clean)
res_driver = md.fit()
print(res_driver.summary())

# 4. Mixed-effects for team
md2 = sm.MixedLM.from_formula("LapTimeSeconds ~ TyreAge", groups="Team", data=all_laps_clean)
res_team = md2.fit()
print(res_team.summary())

# 5. Track-evolution “rubber-in” delta per GP
rubber = []
for gp, grp in all_laps_clean.groupby("GrandPrix"):
    # average lap time on lap 2 vs penultimate lap
    early = grp[grp.LapNumber == 2]["LapTimeSeconds"].mean()
    last  = grp[grp.LapNumber == grp.LapNumber.max() - 1]["LapTimeSeconds"].mean()
    rubber.append({"GrandPrix": gp, "RubberDelta_s": early - last})
rubber_df = pd.DataFrame(rubber)
print(rubber_df)
```

> **Insight**:

```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import GroupKFold, cross_val_score
from sklearn.ensemble import HistGradientBoostingRegressor
import shap

# 1. Feature engineering
all_laps_clean["CumulativeLaps"] = all_laps_clean.groupby(["GrandPrix","Driver"]).cumcount()
best = all_laps_clean.groupby(["GrandPrix","Driver"])["LapTime_s"].transform("min")
all_laps_clean["GapToBest_s"] = all_laps_clean["LapTime_s"] - best

# merge weather if you have it:
# weather = pd.read_csv("weather.csv")  
# all_laps_clean = all_laps_clean.merge(weather, on="GrandPrix", how="left")

# 2. Prepare X, y
cat = ["Driver","Team","Compound"]
num = ["TyreAge","CumulativeLaps","GapToBest_s"]  # + ["TrackTemp","RainFlag"]
ohe = OneHotEncoder(sparse_output=False, drop="first")
X_cat = ohe.fit_transform(all_laps_clean[cat])
X_num = all_laps_clean[num].values
X = np.hstack([X_num, X_cat])
y = all_laps_clean["LapTime_s"].values

# 3. Group-KFold by GrandPrix
gkf = GroupKFold(n_splits=5)
model = HistGradientBoostingRegressor(random_state=42)
scores = cross_val_score(model, X, y, groups=all_laps_clean["GrandPrix"], cv=gkf,
                         scoring="neg_mean_absolute_error")
print("Fold MAEs:", -scores)
print("Mean MAE:", -scores.mean())

# 4. Train final model & explain with SHAP
model.fit(X, y)
# Reconstruct feature names
feature_names = num + list(ohe.get_feature_names_out(cat))  # e.g. ["TyreAge", ..., "Driver_Lec", "Team_Mercedes", ...]

explainer = shap.Explainer(model, X, feature_names=feature_names)
sv = explainer(X, check_additivity=False)

shap.plots.beeswarm(sv)                       # feature importance
shap.plots.partial_dependence(
    "TyreAge",    # the feature name (must match the one in feature_names)
    sv,           # SHAP values (already computed)
    X             # the dataset used for predictions
)
```

```python
# Ensure X_clean and y_clean are not empty before proceeding
if not X_clean.empty and not y_clean.empty:
    # Define preprocessing steps for categorical and numerical features
    # HistGradientBoostingRegressor can handle categorical features natively if they are integer-encoded
    # or by specifying `categorical_features` as a boolean mask or list of indices.
    # For simplicity with ColumnTransformer and broader compatibility, we can use OneHotEncoder.
    # However, let's try OrdinalEncoder first as HGBR can benefit from it,
    # and it avoids high dimensionality of OneHotEncoder with many drivers/teams.

    # Preprocessor for categorical features using OrdinalEncoder
    # Important: HGBR needs to be told these are categorical.
    # We will prepare the data and then tell HGBR which columns are categorical.
    
    # Create a list of boolean flags indicating which features are categorical for HGBR
    # This requires X_clean to be in the final order of features that HGBR will see.
    
    # Let's first define the transformers
    # For numerical features, though HGBR is robust, scaling can sometimes help other models if you switch.
    numerical_transformer = StandardScaler() # Or 'passthrough' if no scaling desired for HGBR

    # For categorical features, we will use OrdinalEncoder.
    # HGBR can then be told these are categorical via its `categorical_features` param.
    categorical_transformer = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)

    preprocessor = ColumnTransformer(
        transformers=[
            ('num', numerical_transformer, numerical_features),
            ('cat', categorical_transformer, categorical_features)
        ], 
        remainder='passthrough' # In case some columns are not in num/cat lists but are in X_clean
    )

    # Define the model
    hgbr_model = HistGradientBoostingRegressor(random_state=42, early_stopping=True)

    # Create the pipeline
    pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                               ('regressor', hgbr_model)])
    
    print("\nModel pipeline created.")

else:
    print("X_clean or y_clean is empty. Cannot proceed with model training.")
```

```python
if not X_clean.empty and not y_clean.empty:
    # Cross-validation setup
    # We group by 'GrandPrix' to ensure that data from the same race doesn't appear in both train and test folds.
    group_kfold = GroupKFold(n_splits=5) # Or choose number of splits, e.g., min(5, number_of_gps)

    print(f"\nStarting cross-validation with {group_kfold.get_n_splits(X_clean, y_clean, groups)} splits...")
    
    # Scoring: Negative Mean Absolute Error (MAE). Scikit-learn's cross_val_score maximizes scores,
    # so error metrics are typically negated. MAE is in the same units as the target.
    # We can also use R-squared ('r2').
    
    try:
        # Before fitting, we need to inform HGBR about categorical features after transformation
        # The ColumnTransformer will output numerical features first, then ordinal encoded categorical features.
        # So, the categorical features will be at the end of the transformed X.
        
        # Get indices of categorical features AFTER ColumnTransformer
        # This is a bit tricky because preprocessor isn't fit yet.
        # An alternative is to fit preprocessor once to see transformed feature names/indices
        # Or, more simply, HistGradientBoostingRegressor can often infer them or be explicitly told
        # via a boolean mask if features are already integer encoded (which OrdinalEncoder does).
        
        # For HGBR with OrdinalEncoder:
        # After OrdinalEncoding, the categorical features are still distinct columns.
        # We need to tell HGBR which of the *transformed* columns are categorical.
        # If numerical_features has N_num columns and categorical_features has N_cat columns,
        # then the categorical ones will be columns N_num to N_num + N_cat - 1 in the output of preprocessor.
        
        # Let's fit the preprocessor to get transformed data and identify categorical indices
        X_transformed_example = preprocessor.fit_transform(X_clean)
        
        # Identify which columns in X_transformed_example correspond to original categorical_features
        cat_feature_indices_in_transformed_X = []
        start_index = len(numerical_features) # Categorical features come after numerical ones
        for i in range(len(categorical_features)):
            cat_feature_indices_in_transformed_X.append(start_index + i)
            
        # Update the regressor step in the pipeline with this info
        pipeline.set_params(regressor__categorical_features=cat_feature_indices_in_transformed_X)
        print(f"Categorical feature indices for HGBR: {cat_feature_indices_in_transformed_X}")

        cv_scores_mae = cross_val_score(pipeline, X_clean, y_clean, groups=groups, cv=group_kfold, scoring='neg_mean_absolute_error')
        cv_scores_r2 = cross_val_score(pipeline, X_clean, y_clean, groups=groups, cv=group_kfold, scoring='r2')

        print(f"\nCross-validation Negative MAE scores: {-cv_scores_mae}")
        print(f"Mean CV MAE: {-cv_scores_mae.mean():.3f} seconds (std: {cv_scores_mae.std():.3f})")
        print(f"Cross-validation R2 scores: {cv_scores_r2}")
        print(f"Mean CV R2: {cv_scores_r2.mean():.3f} (std: {cv_scores_r2.std():.3f})")

        # Fit the final model on all clean data for later use (e.g., feature importance)
        print("\nFitting final model on all clean data...")
        pipeline.fit(X_clean, y_clean)
        print("Final model fitted.")

    except Exception as e:
        print(f"An error occurred during cross-validation or model fitting: {e}")
        import traceback
        traceback.print_exc()
else:
    print("Skipping model training as X_clean or y_clean is empty.")
```

```python
from sklearn.inspection import permutation_importance
from sklearn.utils.validation import check_is_fitted # For a more robust check

# Ensure 'pipeline' was defined and fitting was attempted from the previous cell
if 'pipeline' in locals() and 'X_clean' in globals() and 'y_clean' in globals():
    try:
        # Check if the regressor step in the pipeline is fitted
        check_is_fitted(pipeline.named_steps['regressor'])
        print("\nCalculating Permutation Feature Importance (pipeline's regressor is fitted)...")

        r = permutation_importance(pipeline, X_clean, y_clean,
                                   n_repeats=5, # You can reduce n_repeats for faster execution initially
                                   random_state=42,
                                   scoring='neg_mean_absolute_error') # Use the same scoring as CV

        # Sort features by importance
        # X_clean.columns contains the original feature names before preprocessing
        perm_sorted_idx = r.importances_mean.argsort()

        plt.figure(figsize=(10, max(6, len(X_clean.columns) * 0.5)))
        plt.barh(X_clean.columns[perm_sorted_idx], r.importances_mean[perm_sorted_idx])
        plt.xlabel("Permutation Importance (Impact on MAE)")
        plt.title("Permutation Feature Importance for Lap Time Prediction")
        plt.tight_layout()
        plt.show()

    except NameError: # If pipeline, X_clean or y_clean not defined
        print("Pipeline, X_clean, or y_clean not defined. Skipping feature importance.")
    except AttributeError: # If named_steps['regressor'] doesn't exist (pipeline structure wrong)
        print("Could not find 'regressor' step in the pipeline. Skipping feature importance.")
    except Exception as e: # For other errors, like NotFittedError from check_is_fitted
        print(f"Could not calculate feature importance: {e}")
        print("Ensure the pipeline was successfully fitted in the previous cell.")
else:
    print("Pipeline, X_clean, or y_clean not found in local scope. Skipping feature importance.")
```
